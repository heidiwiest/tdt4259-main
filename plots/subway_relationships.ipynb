```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# convert csvs to dataframes
data = pd.read_csv('../data/training_set_prepped.csv')
```

```python
df = data.loc[(data['neighbourhood_group_cleansed_Brooklyn'] == True) | (data['neighbourhood_group_cleansed_Queens'] == True)]
```

```python
df = df.loc[data['accommodates'] == 1]
```

```python
df = df.loc[df['property_type'] == 20]
```

```python
pd.set_option('display.max_columns', None)
df
```

```python
df['price'] = df['price']**2
```

```python
# Now let's create a scatter plot with hue
plt.figure(figsize=(10, 6))

# Adding a 'location' column to use as a hue which will differentiate Brooklyn and Queens
df['location'] = np.where(df['neighbourhood_group_cleansed_Brooklyn'] == 1, 'Brooklyn', 'Queens')
```

```python
# Create a scatter plot
sns.scatterplot(data=df, x='distance_to_closest_subway', y='price', hue='location')

# Set the plot title and labels
plt.title('Price vs. Subway Distance by Borough')
plt.xlabel('Subway Distance')
plt.ylabel('Price')
plt.legend(title='Location')

# Show the plot
plt.show()
```
